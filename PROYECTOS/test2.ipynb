```python
#Funcion que devuelva el Mayor de dos numeros:
def may(n1,n2):
    if n1 > n2:
        return n1 
    else:
        return n2
num1 = int(input("Escribe el primer numero: "))
num2 = int(input("Escrib el segundo numero:"))
may = max(num1,num2)
print(f"El mayor numero es {may}")
```

```python
#Escriba un programa que pida la anchura y altura de un rectángulo y lo dibuje con caracteres roducto (*):# 
def rectangulo(ancho,altura):
    for i in range(1,altura+1):
        for j in range (1,ancho+1):
            print("*",end="")
        print()
  
ancho = int(input("anchura del rectangulo:"))
alto = int(input("Altura del rectangulo: "))
rectangulo(ancho,alto)
```

```python
#Escriba un programa que pida la anchura y altura de un rectángulo y lo dibuje con caracteres roducto (*):# 
def rectangulo_personalizado(ancho,altura,caracter):
    for i in range(1,altura+1):
        for j in range (1,ancho+1):
            print(caracter,end="")
        print()
  
ancho = int(input("anchura del rectangulo:"))
alto = int(input("Altura del rectangulo: "))
caracter = input("Caracter a imprimir: ")
rectangulo_personalizado(ancho,alto,caracter)
```

```python
#Escribe un programa que pida la anchura de un triangulo y lo dibuje con caracter del producto (*)#
def triangulo(acho):
    for i in range(1,ancho+1):
        for j in range(1,i+1):
            print("*",end="")  
        print()

ancho = int(input("Anchura del triangulo: "))
triangulo(ancho)
triangulo(ancho)
```

```python
#Realiza una funcion que calcule el factorial de un numero, sabiendo que:# 
# numero = n * (n-1)*(n-2)* 1#
#Ejemplo 5/ = 5 x 4 x 3 x 2 x 1 = 120#
# 0/ = 1  

def factorial(numero):
    f = 1 
    for i in range(1,numero+1):
        f = f * i
    return f

n = int(input("Escribe un numero:"))
resultado = factorial(n)
print(f"El factorial es: {resultado}")
```

```python
def factorial(numero):
    f = 1 
    for i in range(1,numero+1):
        f = f + i
    return f 

n = int(input("Escribe un numero:"))
resultado = factorial(n)
print(f"El factorial es: {resultado}")
```
