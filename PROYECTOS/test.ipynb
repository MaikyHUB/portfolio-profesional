```python
# Pedirle al usuario una cantidad, si esa cantidad es mayor a 5000, aplicarle un descuento del 10%. Mostrar cunato tndria que 
cantidad = int(input("Ingrese la cantidad: "))
if cantidad > 5000:
    resultado = cantidad-(cantidad*10/100)
    print(f"Usted tiene que pagar{cantidad}")
else:
    resultado = cantidad
print(f"Usted tiene que pagar {resultado} no se le aplico el descueto de el 10%")
 
```

```python
# Usted tiene una Nota de 0 - 10 si es menor que 5= insufisiene 5= suficiente 6= bien 7- 8 notable 9 10 sobresaliente. Si no esta entre 0 - 10 nota no valida.
nota = float(input("Tu nota es:"))
if nota < 0 or nota > 10:
    print("nota no valida")
elif nota < 5 :
    print( "insufisiente")
elif nota < 6 :
    print( "sufisiente")
elif nota < 7 :
    print( "bien")
elif nota < 9 :
    print( "notable")  
else:
    print("sobresaliente")
```

```python
#pide una opcion tiene que estar entre 1 - 6 
#1-suma 
#2-resta 
#3-producto
#4-division
#5-resto de la division
#6-potencia

print("1 suma ")
print("2 resta ")
print("3 producto ")
print("4 division ")
print("5 reso de la division ")
print("6 potencia ")

opcion = int(input("elige la opcion:"))
n1 = int(input("elige el numero uno: "))
n2 = int(input("elige el numero dos: "))

if opcion == 1:
    print(f" Suma: = {n1 + n2}")
elif opcion == 2:
    print(f" Resta: = {n1 - n2}")
elif opcion == 3:
    print(f" Producto: = {n1 * n2}")
elif opcion == 4:
    print(f" Division: = {n1 / n2} ")
elif opcion == 5:
    print(f"resto de la division: = {n1 // n2}")
elif opcion == 6:
    print(f" Potencia: = {n1 ** n2}")
else:
    print("Opcion invalida")

    
```
