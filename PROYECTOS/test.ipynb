```python
import random

secreto = random.randint(1, 20)
encontrado = False

print("¡Adivina el número secreto entre 1 y 20!")

while not encontrado:
    try:
        num = int(input("Introduce el número: "))
    except ValueError:
        print("Entrada no válida. Por favor, introduce un número entero.")
        continue

    if num == secreto: 
        encontrado = True
    elif num > secreto:
        print("El número introducido es mayor. Intenta de nuevo.")
    else:
        print("El número introducido es menor. Intenta de nuevo.")

print(f"¡Felicidades! Adivinaste el número secreto, que era {secreto}.")
```

```python
print(f"1. Cacular el espacio libre: ")
print(f"2. MB -> GB")
print(f"3. TB -> GB")
print(f"4. Saliendo del programa")
opcion = int(input("Introduce una opcion: "))
if opcion == 1: 
    espacio_libre = int(input("Introduce el epacio libre: "))
    espacio_ocupado = int(input("Introduce el espacio ocupado: "))
    print(f"Resultado: {espacio_ocupado-espacio_libre}")
elif opcion == 2:
    mb = int(input("Introduce los megas: "))
    print(f"Resultado: {mb/1024} GB")
elif opcion == 3:
    tb = int(input("Introduce los teras: "))
    print(f"Resultado: {tb*1024} GB")
elif opcion == 4:
    print(f"Saliendo del programa")
else:
    print(f"Opcion invalida")
```

```python
print ("---Herrameintas de administracion ---")
print (f"1. Clcular el porcentaje de uso de la cpu")
print (f"2. Calcular el tiempo estimado de copia de seguridad")
print (f"3. Comprobar si una contraseña es segura ")
print (f"4. Salir")
opcion = int(input("Seleciona una opcion:"))
if opcion == 1:
    capacidad = float(input("Introduce la capacidad total de la CPU (GHz): "))
    uso_actual = float(input("Introduce el usuo actual de la CPU (GHz): "))
    print (f"el porcentaje de uso de CPU es: {capacidad * uso_actual / 100:.2f} %")
elif opcion == 2:
    tamaño = float(input("Introduce el tamaño de los datos (GB): "))
    velocidad = float(input("Introduce la velocidad de transferencia (Gb/min): "))
    print (f"El tiempo estimado de copia de seguridad es: {tamaño / velocidad:.2f} minutos")
elif opcion == 3:
    contrasenna = input("Introduce la contraseña: ")
    if len(contrasenna) >= 8:    
        print(f"Contraseña segura ")
    else:
        print(f"Contraseña debil")
elif opcion == 4:
    print(f"Finalizando herramienta de administracion...")
else:
    print (f"Opcion incorrecta. Intente nuevamente")
```
